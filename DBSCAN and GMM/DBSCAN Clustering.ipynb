# 3) use K-means, agglomerative, dbscan


```python
kmean = KMeans(n_clusters=2, init='k-means++')
kmean.fit(dataset1.iloc[: ,2:4]) 

kmeans_inertia = []
kmeans_inertia.append(kmean.inertia_)

agg = AgglomerativeClustering(n_clusters=2)
agg.fit((dataset1.iloc[: ,2:4]))

dbscan = DBSCAN(eps=0.2, min_samples=2)
dbscan.fit((dataset1.iloc[: ,2:4]))
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].scatter(dataset1.iloc[: ,2], dataset1.iloc[: ,3], c = kmean.labels_)
ax[1].scatter(dataset1.iloc[: ,2], dataset1.iloc[: ,3], c = agg.labels_)
ax[2].scatter(dataset1.iloc[: ,2], dataset1.iloc[: ,3], c = dbscan.labels_)
```
