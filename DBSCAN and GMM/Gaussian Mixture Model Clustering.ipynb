# Import Libraries

```python
#!pip install pandas plotnine
# use any plotting library seaborn matplotlib,..........
# Plotting
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
import pandas as pd
import seaborn as sns; sns.set()
from plotnine import *
# Sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_circles, make_moons, make_blobs
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

# Load Datasets

```python
n_samples = 1500
dataset1 = make_circles(n_samples=n_samples, factor=.5, noise=.05)
dataset1 = pd.DataFrame(dataset1[0])

dataset2 = make_moons(n_samples=n_samples, noise=.05)
dataset2 = pd.DataFrame(dataset2[0])

dataset3 = make_blobs(n_samples=n_samples, random_state=8)
dataset3 = pd.DataFrame(dataset3[0])

dataset4 = np.random.rand(n_samples, 2), None
dataset4 = pd.DataFrame(dataset4[0])

dataset5 = pd.read_csv("dataset5.csv")
dataset6 = pd.read_csv("dataset6.csv")
dataset7 = pd.read_csv("dataset7.csv")
dataset8 = pd.read_csv("dataset8.csv")
dataset9 = pd.read_csv("dataset9.csv")
```

# Tasks
1) Use standard scalar
2) Use Gaussian mixture model for clustering
3) Plot dataset after clustering

# 1) Use standard scalar

```python
z = StandardScaler()

dataset1[["x","y"]] = z.fit_transform(dataset1)
dataset2[["x","y"]] = z.fit_transform(dataset2)
dataset3[["x","y"]] = z.fit_transform(dataset3)
dataset4[["x","y"]] = z.fit_transform(dataset4)
dataset5[["x","y"]] = z.fit_transform(dataset5)
dataset6[["x","y"]] = z.fit_transform(dataset6)
dataset7[["x","y"]] = z.fit_transform(dataset7)
dataset8[["x","y"]] = z.fit_transform(dataset8)
dataset9[["x","y"]] = z.fit_transform(dataset9)
```

**Plot Cluster Function**

```python
def plot_clusters(dataset, kmeans_labels, gmm_labels, probs):
    size = 50 * probs.max(1) ** 2  # square emphasizes differences
    
    fig, ax = plt.subplots(1,3, figsize=(10,2))
    ax[0].scatter(dataset['x'], dataset['y'])
    ax[1].scatter(dataset['x'], dataset['y'], c = kmeans_labels)
    ax[2].scatter(dataset['x'], dataset['y'], c = gmm_labels, s=size)

    ax[0].set_xlabel('X')
    ax[1].set_xlabel('X')
    ax[2].set_xlabel('X')

    ax[0].set_ylabel('Y')

    ax[0].set_title('Data Before Clustering')
    ax[1].set_title('Data After Clustering(KMeans)')
    ax[2].set_title('Data After Clustering(GMM)')
    
    plt.show()
```

# 2) Use Gaussian mixture model for clustering
# 3) Plot dataset after clustering

```python
gmm = GaussianMixture(n_components=2, random_state=0).fit(dataset1)
labels = gmm.predict(dataset1)
probs = gmm.predict_proba(dataset1)

kmean = KMeans(n_clusters=2, init='k-means++')
kmean.fit(dataset1.iloc[: ,2:4]) 

kmeans_inertia = []
kmeans_inertia.append(kmean.inertia_)
# Plot Clusters
plot_clusters(dataset1, kmean.labels_, labels, probs)
```
